```python
#Definimos una función Horner con los (coeffs) coeficientes del polinomio de mayor a menor grado, (x0) punto que se va a 
# evaluar el polinomio y su derivada
def d_horner(coeffs, x0):
    #Grado del polinomio
    n = len(coeffs) - 1 
    # p(x0) inicializado de mayor a menor grado
    p = coeffs[0] 
    # dp(x0) empieza en cero
    dp = 0  
    #Creamos un for de 1 a n+1
    for i in range(1, n + 1):
        #Actualiza a nuevo dp
        dp = dp * x0 + p  
        #Actualiza a p(x0)
        p = p * x0 + coeffs[i]
    # Me regresa p(x0) y dp(x0)
    return p, dp

# Ejemplo
coeffs = [1, -3, 2, -1]  # Representa el polinomio p(x) = x^3 - 3x^2 + 2x - 1
x0 = 2
p_x0, dp_x0 = d_horner(coeffs, x0)

print(f"p({x0}) = {p_x0}")
print(f"p'({x0}) = {dp_x0}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import hermite, factorial

#a) Probabilidad clásica y el tiempo en intervalo
def prob_clasica(x,x0,T):
    return 1 / (T * np.sqrt(x0**2 - x**2))

def tiempo_clasico(x, dx, x0, T):
    P_c = prob_clasica(x,x0,T)
    return P_c *dx #Tiempo P_c(x)*dx

#b) Funciones de onda cuánticas
def quantum_wavefunction(n,x):
    Hn= hermite(n) #Polinomio de Hermite
    normalizacion=1.0/np.sqrt(2**n * factorial(n)* np.sqrt(np.pi)) #Constante de normalizacion
    return normalizacion *Hn(x)* np.exp(-x**2/2) #Función de onda

#Parámetros
x=np.linspace(-0.99,0.99,1000)
x0=1
omega=2*np.pi
T=2*np.pi/omega
dx=0.01

#a) Calculamos P_c(x) y t
P_c=prob_clasica(x,x0,T)
delta_t= tiempo_clasico(x,dx,x0,T)

#b) Obteneemos los cuadrados de las funciones de onda
nvals=[3,10,20,150]
psi_cuadrado = {n: quantum_wavefunction(n,x)**2 for n in nvals}

#Gráficas
#a) Probabilidad clásica
plt.figure(figsize=(10,6))
plt.plot(x,P_c, label="Probabilidad clásica $P_c(x)$",color="blue")
plt.xlabel("$x$")
plt.ylabel("$P_c(x)$")
plt.title("Probabilidad clásica del oscilador")
plt.grid(True)
plt.legend()
plt.show()

#a) Tiempo en intervalo dx
plt.figure(figsize=(10,6))
plt.plot(x,delta_t,label=f"Tiempo en intervalo $dx={dx}$", color="green")
plt.xlabel("$x$")
plt.ylabel("$\Delta t$")
plt.title("Tiempo en un intervalo dx para el oscilador clásico")
plt.grid(True)
plt.legend()
plt.show()

#b) Funciones de onda cuánticas
plt.figure(figsize=(10,6))
for n in nvals:
    plt.plot(x,psi_cuadrado[n], label=f"$|\psi_{n}(x)|^2$, n={n}")
plt.xlabel("$x$")
plt.ylabel("$|\psi_n(x)|^2$")
plt.title("Cuadrados de las funciones de onda cuánticas")
plt.grid(True)
plt.legend()
plt.show()
```

```python
from math import sqrt, pi, factorial, exp 

def hermite(n, x):
    #Calcula el polinomio de Hermite H_n(x) y su derivada H'_n(x).
    val0 = 1.0  # H_0(x)
    val1 = 2 * x  # H_1(x)
    
    for j in range(1, n):  # Recurrencia para H_n(x)
        val2 = 2 * x * val1 - 2 * j * val0
        val0, val1 = val1, val2  # Actualizar valores anteriores
    
    dval2 = 2 * n * val0  # Derivada de H_n(x)
    return val2, dval2

def psiqho(x, n):
    #Calcula la función de onda unidimensional del oscilador armónico cuántico.
    mohbar = 1.0  # mω/ħ, definido como 1 en unidades naturales
    al = 1.0  # α, definido como 1 en unidades naturales

    # Parte gaussiana de la función de onda
    psival = (mohbar ** 0.25) * exp(-0.5 * al * mohbar * x**2)
    
    # Multiplicar por el polinomio de Hermite correspondiente
    psival *= hermite(n, sqrt(mohbar) * x)[0]
    
    # Normalización
    psival /= sqrt(2**n * factorial(n) * sqrt(pi)) 
    return psival

def kinetic(psi, x, nametoval):
    # Calcula la energía cinética usando diferencias finitas en la posición x.
    hom = 1.0  # ħω, definido como 1 en unidades naturales
    h = 0.005  # Paso de discretización

    # Evaluar ψ(x), ψ(x + h) y ψ(x - h)
    psiold = psi(x, nametoval)
    psip = psi(x + h, nametoval)
    psim = psi(x - h, nametoval)

    # Laplaciano usando diferencias finitas
    lapl = (psip + psim - 2.0 * psiold) / h**2 

    # Energía cinética
    kin = -0.5 * hom * lapl / psiold
    return kin

# Números cuánticos para cada dirección
nx = 2
ny = 3
nz = 5

# Calcular la energía total tridimensional
ener = (kinetic(psiqho, 1, nx) + 
        kinetic(psiqho, 1, ny) + 
        kinetic(psiqho, 1, nz))

# Imprimir el resultado
print(f"Energía total: {ener:}")
```

```python
from math import sqrt, pi, factorial, exp
import cmath

#Polinomios de Hermite
def hermite(n, x):
    valor0 = 1.0
    valor1 = 2 * x
    for j in range(1, n):
        valor2 = 2 * x * valor1 - 2 * j * valor0
        valor0, valor1 = valor1, valor2
    derivada2 = 2 * n * valor0
    return valor2, derivada2

#Función de onda del oscilador armónico cuántico
def psi_oscilador(x, nombre):
    n = nombre["n"]
    mw_h = nombre["mw/h"]
    alfa = nombre["alfa"]
    psi_valor = mw_h**0.25 * exp(-0.5 * alfa * mw_h * x**2)
    psi_valor *= hermite(n, sqrt(mw_h) * x)[0]
    psi_valor /= sqrt(2**n * factorial(n) * sqrt(pi))
    return psi_valor

# Función de onda de una partícula en una caja infinita
def psi_caja(x, nombre):
    n = nombre["n"]
    longitud_caja = nombre["longitud_caja"]
    return cmath.exp(2 * pi * n * x * 1j / longitud_caja) / sqrt(longitud_caja)

#Energía cinética: Versión original
def energia_cinetica_original(psi, x, nombre, h=0.005):
    hbarra = 1.0
    psi_menos = psi(x - h, nombre)
    psi_cero = psi(x, nombre)
    psi_mas = psi(x + h, nombre)
    laplaciano = (psi_mas + psi_menos - 2 * psi_cero) / h**2
    energia = -0.5 * hbarra * laplaciano / psi_cero
    return energia

#Energía cinética: Versión mejorada
def energia_cinetica_mejorada(psi, x, nombre, h=0.005):
    hbarra = 1.0
    psi_menos2 = psi(x - 2 * h, nombre)
    psi_menos1 = psi(x - h, nombre)
    psi_cero = psi(x, nombre)
    psi_mas1 = psi(x + h, nombre)
    psi_mas2 = psi(x + 2 * h, nombre)
    laplaciano = (-psi_mas2 + 16 * psi_mas1 - 30 * psi_cero + 16 * psi_menos1 - psi_menos2) / (12 * h**2)
    energia = -0.5 * hbarra * laplaciano / psi_cero
    return energia

#Prueba de comparación
def prueba_comparacion_energia():
    x = 1.0  # Punto de evaluación
    pasos_h = [10**(-i) for i in range(1, 6)]  # Valores de h para probar

    # Parámetros para el oscilador armónico cuántico
    parametros_oscilador = {"n": 100, "mw/h": 1.0, "alfa": 1.0}
    # Parámetros para la partícula en la caja
    parametros_caja = {"n": 2, "longitud_caja": 2 * pi}

    # Energía cinética con la versión original
    oscilador_original = [energia_cinetica_original(psi_oscilador, x, parametros_oscilador, h) for h in pasos_h]
    caja_original = [energia_cinetica_original(psi_caja, x, parametros_caja, h) for h in pasos_h]

    # Energía cinética con la versión mejorada
    oscilador_mejorado = [energia_cinetica_mejorada(psi_oscilador, x, parametros_oscilador, h) for h in pasos_h]
    caja_mejorada = [energia_cinetica_mejorada(psi_caja, x, parametros_caja, h) for h in pasos_h]

    # Comparar resultados
    formato_fila = "{:0.1e} {:1.16f} {:1.16f} {:1.16f} {:1.16f}"
    print("h          Oscilador Original    Oscilador Mejorado    Caja Original        Caja Mejorada")
    for h, o_o, o_m, c_o, c_m in zip(pasos_h, oscilador_original, oscilador_mejorado, caja_original, caja_mejorada):
        print(formato_fila.format(h, o_o, o_m, c_o, c_m))

if __name__ == "__main__":
    prueba_comparacion_energia()
```

```python
# Definimos la función de onda para el potencial escalón en diferentes regiones.

def psistep1(x, params):
    #Función de onda en la región 1 para un potencial escalón.

    A = params["A"]
    B = params["B"]
    ik1 = params["ik1"]
    psi = A * exp(ik1 * x) + B * exp(-ik1 * x)
    return psi

def psistep2(x, params):
    #Función de onda en la región 2 para un potencial escalón.
    C = params["C"]
    k2 = params["k2"]
    psi = C * exp(-k2 * x)
    return psi
```
